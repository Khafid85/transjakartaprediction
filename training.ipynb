### **Domain Knowledge**

**apa itu LRT**

LRT (Light Rail Transit) adalah sistem transportasi massal berbasis rel yang beroperasi di atas permukaan tanah, tetapi dapat juga beroperasi melalui jalur bawah tanah. LRT dirancang untuk mengangkut penumpang dengan jumlah sedang hingga besar pada jarak yang relatif pendek hingga menengah di area perkotaan. LRT memiliki kecepatan yang lebih tinggi daripada bus tetapi lebih lambat jika dibandingkan dengan kereta api. LRT menggunakan tenaga listrik, sehingga lebih ramah lingkungan dibandingkan dengan moda transportasi berbasis bahan bakar fosil. LRT banyak digunakan di kota-kota besar di seluruh dunia sebagai bagian dari upaya mengurangi kemacetan lalu lintas, mengurangi emisi karbon, dan menyediakan transportasi yang efisien dan nyaman bagi warga.

![LRT](https://awsimages.detik.net.id/community/media/visual/2023/08/28/asyik-lrt-jabodebek-beroperasi-penuh-siang-ini-1_169.jpeg?w=1200)


**apa itu MRT**

MRT (Mass Rapid Transit) adalah sistem transportasi kereta api perkotaan yang dirancang untuk mengangkut penumpang dalam jumlah besar secara cepat dan efisien di wilayah metropolitan. MRT biasanya beroperasi di jalur yang eksklusif, terpisah dari lalu lintas lain, dan sering kali menggunakan kombinasi jalur layang, permukaan, dan bawah tanah.

![mrt](https://upload.wikimedia.org/wikipedia/commons/3/34/MRTset3.jpg)

**perbedaan LRT dan MRT**

Kedua sistem ini sering ditemukan beroperasi bersamaan dalam jaringan transportasi yang sama, saling melengkapi untuk memberikan cakupan yang luas dan pilihan yang efisien bagi penumpang di kota besar. Berikut ini adalah 5 jenis perbedaan antara LRT dan MRT

**1. Kapasitas dan Skala**
- **MRT**: Dirancang untuk menangani volume penumpang yang sangat tinggi dengan kapasitas yang besar, membuatnya ideal untuk kota-kota besar dengan kepadatan penduduk tinggi.
- **LRT**: Kapasitasnya lebih rendah dibandingkan MRT, lebih cocok untuk jumlah penumpang sedang. LRT sering digunakan di kota-kota yang lebih kecil atau untuk melengkapi sistem transportasi yang ada di kota besar.

**2. Kecepatan**
- **MRT**: Umumnya lebih cepat daripada LRT karena beroperasi di jalur eksklusif tanpa harus berbagi dengan lalu lintas kendaraan lain.
- **LRT**: Kecepatannya lebih variatif, tergantung pada apakah ia beroperasi di jalur khusus atau berbagi jalur dengan lalu lintas lain.

**4. Biaya Pembangunan**
- **MRT**: Memerlukan investasi yang lebih besar dalam hal konstruksi dan pemeliharaan karena infrastruktur yang lebih kompleks, seperti stasiun bawah tanah dan peralatan yang canggih.
- **LRT**: Lebih ekonomis dari segi biaya pembangunan dan pemeliharaan karena memiliki infrastruktur yang lebih sederhana.

**5. Lingkup Penggunaan**
- **MRT**: Cocok untuk mengangkut jumlah penumpang besar antar wilayah dalam kota atau antar kota, menghubungkan titik-titik penting seperti pusat bisnis dan residensial.
- **LRT**: Lebih sering digunakan untuk penghubung pada jarak menengah dan dapat beroperasi di daerah pinggiran atau antara kota.

### **Data Understanding & Data Cleaning & Data Transformation**

pada proses ini dilakukan beberapa tahapan yaitu :

1. Join data
- Pada proses ini dilakukan joining seluruh sumber data dengan key yaitu bulan dan tahun dengan menggunakan pandas dengan metode full join agar tidak terdapat data yang hilang
2. Sorting data
- sorting data dilakukan dari data terlama dengan menggunakan kolom sintetis date dengan membuat tanggal default 1 untuk setiap bulannya, untuk mendapatkan format data yang sesuai dengan pandas
3. pengisian missing value kolom LRT dan MRT 
- berdasarkan kondisi real, LRT dan MRT baru beroperasi sejak tahun 2019, oleh karenanya sebelum tahun tersebut datanya diisi dengan nilai null

```python

def join_data(dfs, kolom_acuan):
    #slicing data pertama
    hasil = dfs[0]

    for df in dfs[1:]:
        #dfs[1:] berarti slicing data pertama hingga terakhir
        hasil = pd.merge(hasil, df, on=kolom_acuan, how='outer')
    
    return hasil
```

```python
dfs = [data1, data2, data3, data4, data5, data6]
kolom = ['bulan', 'tahun']
final_result = join_data(dfs=dfs, kolom_acuan=kolom)
```

### LRT Pertama kali beroperasi di Jakarta

LRT (Light Rail Transit) Jakarta mulai beroperasi secara komersial pada tanggal 1 Desember 2019. berarti sebelum desember 2019, jumlah penumpang dan jumlah armada LRT akan dianggap nol sebelum tanggal ini.

```python
def ganti_lrt(final_result):
    #kondisi di tahun 2015 - 2018 seluruh data armada & perjalanan lrt = 0
    tahun = [2015, 2016, 2017, 2018]
    for j in tahun:
        final_result.loc[final_result['tahun'] == j, 'jumlah_penumpang_lrt'] = 0
        final_result.loc[final_result['tahun'] == j, 'jumlah_perjalanan_lrt'] = 0
    
    #kondisi di tahun 2019 pada bulan 1 - 11 seluruh armada & perjalanan lrt = 0
    for i in range(1, 12):
        final_result.loc[(final_result['tahun'] == 2019) & (final_result['bulan'] == i), 'jumlah_penumpang_lrt'] = 0
        final_result.loc[(final_result['tahun'] == 2019) & (final_result['bulan'] == i), 'jumlah_perjalanan_lrt'] = 0
    
    return final_result
```

```python
final_result = ganti_lrt(final_result)
```

Alasan untuk ini adalah bahwa uji normalitas memerlukan data yang lengkap untuk menghitung statistik tes dan p-value dengan tepat. Missing values dapat menyebabkan bias dalam perhitungan dan bisa mengganggu interpretasi hasil uji.

```python
def mengisi_missing_value(df, columns):
    for column in columns:
        data = df[column].dropna()
        stat, p_value = shapiro(data)
        print(f'Column: {column}, Shapiro-Wilk p-value: {p_value}')

        #jika p value diatas 0.05 data diisi dengan mean karena berdistribusi normal
        if p_value > 0.05:
            fill_value = data.mean()
            print(f'Filling missing values in {column} with mean: {fill_value}')
            print('data berdistribusi normal')
            print('='*100)
        else:

        #jika p value dibawah 0.05 data diisi dengan median karena tidak berdistribusi normal
            fill_value = data.median()
            print(f'berikut ini adalah nilai yg mengimputasi data : {fill_value}')
        #jika median = 0 maka data akan diisi dengan mean agar nilainya lebih realistis
            if fill_value != 0:
                fill_value = data.median()
                print(f'Filling missing values in {column} with median: {fill_value}')
            else:
                fill_value = data.mean()
                print(f'dikarenakan nilai yang mengimputasi data adalah 0 maka data akan diisi dengan mean agar nilainya lebih realistis, berikut nilai yang mengimputasi data : {fill_value}')
            print('data tidak berdistribusi normal')
            print('='*100)
        df[column].fillna(fill_value, inplace=True)

    return df
```

Membuat fungsi untuk deteksi dan visualisasi outliers

```python
def deteksi_dan_visualisasi_outlier(df):
    def hitung_outlier(col):
        q1 = np.percentile(col, 25)
        q3 = np.percentile(col, 75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        return col[(col < lower_bound) | (col > upper_bound)]

    # Agar kolom bulan dan tahun tidak masuk ke dalam output
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    filtered_cols = []
    for col in numeric_cols:
        if col not in ['bulan', 'tahun']:
            filtered_cols.append(col)
    numeric_cols = filtered_cols
    n_cols = len(numeric_cols)
    n_rows = (n_cols + 1) // 2
    
    fig, axs = plt.subplots(n_rows, 2, figsize=(20, 5 * n_rows))
    axs = axs.ravel()

    for i, col_name in enumerate(numeric_cols):
        outliers = hitung_outlier(df[col_name])
        if not outliers.empty:
            sorted_outliers = outliers.sort_values() 
            print(f"-> {col_name} memiliki {len(outliers)} outlier, dengan nilai sebagai berikut: {sorted_outliers.values}")
        else:
            print(f"-> {col_name} tidak memiliki outlier.")
        sns.boxplot(x=df[col_name], ax=axs[i])
        axs[i].set_title(f'Boxplot dari {col_name}', fontsize=14)
        axs[i].set_xlabel(col_name, fontsize=12)
        axs[i].set_ylabel('Values', fontsize=12)
        axs[i].ticklabel_format(style='plain', axis='x')

    plt.tight_layout()
    plt.show()
```

```python
def calculate_rmse_differences(df):
    results = {}
    col_names = list(df.columns)
    
    for i in range(0, len(col_names), 2):
        training_col = col_names[i]
        testing_col = col_names[i+1]
        
        # Ambil nama algoritma dari kolom training atau testing
        algo_name = training_col.replace(' Training', '')
        
        # Hitung selisih RMSE
        rmse_training = df[training_col].values[0]
        rmse_testing = df[testing_col].values[0]
        rmse_difference = abs(rmse_training - rmse_testing)
        
        # Simpan hasil
        results[algo_name + ' RMSE Difference'] = rmse_difference

    return pd.DataFrame(results, index=["RMSE Difference"])

# Menggunakan fungsi
try:
    rmse_diffs = calculate_rmse_differences(results.T)
    rmse_diffs
except ValueError as e:
    print(e)
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
def prepare_and_evaluate_forecasting(data, models, number_of_lags):
    data.set_index('date', inplace=True)
    split_point = int(0.8 * len(data))
    train = data.iloc[:split_point]
    test = data.iloc[split_point:]
    y_train = train['jumlah_penumpang_tj']
    y_test = test['jumlah_penumpang_tj']

    def create_features_from_lags(data, lags):
        X = pd.DataFrame(index=data.index)
        y = data.copy()
        for lag in range(1, lags + 1):
            X[f'lag_{lag}'] = data.shift(lag)
        X = X.dropna()
        y = y.loc[X.index]
        return X, y

    X_train, y_train_adjusted = create_features_from_lags(y_train, number_of_lags)
    X_test, y_test_adjusted = create_features_from_lags(y_test, number_of_lags)

    trained_models = {}
    results = pd.DataFrame()

    for name, model in models.items():
        model.fit(X_train, y_train_adjusted)
        yhat = model.predict(X_test)
        
        r2 = r2_score(y_test_adjusted, yhat)
        rmse = np.sqrt(mean_squared_error(y_test_adjusted, yhat))
        results.loc[name, 'R2'] = r2
        results.loc[name, 'RMSE'] = rmse

        trained_models[name] = model

        plt.figure(figsize=(10, 4))
        plt.plot(y_test_adjusted.index, y_test_adjusted, label='Actual Data', color='blue')
        plt.plot(y_test_adjusted.index, yhat, label=f'Forecast by {name}', linestyle='--', color='red')
        plt.title(f'Time Series Forecasting using {name}')
        plt.xlabel('Time')
        plt.ylabel('Value')
        plt.legend()
        plt.show()

    return results, trained_models


# Define models
models = {
    "Linear Regression": LinearRegression(),
    "Ridge": Ridge(),
    "Lasso": Lasso(),
    "ElasticNet": ElasticNet(),
    "Decision Tree Regressor": DecisionTreeRegressor(),
    "KNN": KNeighborsRegressor(),
    "Random Forest Regressor": RandomForestRegressor(),
    "AdaBoost Regressor": AdaBoostRegressor(),
    "Gradient Boosting": GradientBoostingRegressor(),
    "XGB": XGBRegressor()
}

results, trained_models = prepare_and_evaluate_forecasting(training_time_series, models, 3)

# Cari model dengan RMSE terendah
best_model_name = results['RMSE'].idxmin()
best_model = trained_models[best_model_name]
results
```

```python
def random_forest_time_series_tuning(y, lags_count):

    def create_lagged_features(series, lags):
        df = pd.DataFrame(series)
        for lag in range(1, lags + 1):
            df[f'lag_{lag}'] = series.shift(lag)
        df.dropna(inplace=True)  # Drop rows with NaN values created by shifting
        return df.iloc[:, 1:], df.iloc[:, 0]  # X, y

    X_train, y_train = create_lagged_features(y, lags_count)

    rf = RandomForestRegressor(random_state=42)

    param_dist = {
        'n_estimators': [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000],
        'max_depth': np.arange(1, 30, 1),
        'min_samples_split': np.arange(2, 20, 1),
        'min_samples_leaf': np.arange(1, 20, 1),
        'max_features': ['sqrt', 'log2']
    }

    # def rmse_scorer(estimator, X, y):
    #     y_pred = estimator.predict(X)
    #     mse = mean_squared_error(y, y_pred)
    #     rmse = math.sqrt(mse)
    #     return rmse 

    # rmse_custom_scorer = make_scorer(rmse_scorer, greater_is_better=False)

    random_search = RandomizedSearchCV(
        estimator=rf,
        param_distributions=param_dist,
        n_iter=300,
        cv=5,
        n_jobs=-1,
        verbose=2,
        scoring='r2',
        random_state=42
    )

    random_search.fit(X_train, y_train)

    best_score = random_search.best_score_ 
    best_params = random_search.best_params_
    best_model = random_search.best_estimator_

    return best_params, best_model, best_score
```
